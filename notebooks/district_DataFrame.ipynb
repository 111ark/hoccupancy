## District Data Frame
Create a DataFrame with all relevant data from the district. We will need this to assign a hospital to each district for the mapping and elimination process.

```python
# Imports
import pandas as pd
import json
```

```python
with open('../data/raw/ch-districts.geojson') as f:
    district_data = json.load(f)
```

```python
# extract property data from the list of dicts
property_data = [d['properties'] for d in district_data['features']]

district_df = pd.DataFrame(property_data)
district_df.head()
```
