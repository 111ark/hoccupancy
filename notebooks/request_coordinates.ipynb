```python
import pandas as pd
import numpy as np
import requests
import os



import time
import random
```

```python
def get_location_coordinates(query):
    url = "https://nominatim.openstreetmap.org/search"
    headers = {
        'User-Agent': 'Mozilla/5.0 (compatible; Python Script)'
    }
    params = {
        'q': query,
        'format': 'json',
        'polygon': 1,
        'addressdetails': 1
    }
    response = requests.get(url, params=params, headers=headers)
    
    # Check if the response contains data
    if response.status_code == 200 and response.json():
        data = response.json()[0]  # Get the first result
        latitude = data['lat']
        longitude = data['lon']
        return latitude, longitude
    else:
        return None, None

# Example usage
latitude, longitude = get_location_coordinates("135 Pilkington Avenue, Birmingham")
print(f"Latitude: {latitude}, Longitude: {longitude}")
```

```python
raw_data = pd.read_excel('../data/raw/kzp22_data.xlsx', sheet_name='KZ2022_KZP22', engine='openpyxl')
adr_data = raw_data[["KT", "Inst", "Adr", "Ort", "Typ"]]
```

```python
adr_data['Full_adr'] = adr_data['Adr'].astype(str) + ' ' + adr_data['Ort'].astype(str)


```

```python

coords = []

for address in adr_data['Full_adr']:
    lat, lon = get_location_coordinates(address)
    # coords = [lat, lon]
    coords.append((lat, lon))
    # coords
    time.sleep(random.uniform(0.1, 1.1))

adr_data['coords'] = coords
    
print("ready")

```
